## Task2

### Q2: Write a python program to scrape data for “Data Scientist”  Job position in“Bangalore” location. You have to scrape the job-title, job-location, company_name, full job-description. You have to scrape first 10 jobs data.

```python
import pandas as pd
import selenium
import time
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
```

```python
driver=webdriver.Chrome(r'C:\Users\home\Downloads\chromedriver_win32\chromedriver.exe')
```

```python
#get url 
url='https://www.naukri.com/'
driver.get(url)
```

```python
#seaching job name in seachbar
search_job=driver.find_elements_by_id('qsb-keyword-sugg')[0]


#sending keys to find jobs in seach bar
search_job.send_keys('Data Scientist')

#searching location in searchbar
search_loc=driver.find_elements_by_id('qsb-location-sugg')[0]

#sending keys to find job location in seachbar
search_loc.send_keys('Bangalore')

#xpath to seach button
search_btn=driver.find_element_by_xpath("//button[@class='btn']")
#clicking seach button
search_btn.click()
```

```python
#get urls tags
url_tags=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
```

```python
# fetching urls to open job link
urls=[]
for i in url_tags:
    urls.append(i.get_attribute("href"))
urls
```

```python
# defining the variables
job_title=[]
job_location=[]
company_name=[]
full_job_description=[]
```

```python
for i in urls[:10]:
    driver.get(i)
    time.sleep(3)
    # fetching job titles
    try:
        job=driver.find_element_by_xpath('//h1[@class="jd-header-title"]')
        job_title.append(job.text)
    except NoSuchElementException:
        job_title.append('-')
    
    #fetching job location
    try:
        loc=driver.find_element_by_xpath('//div[@class="loc"]/span')
        job_location.append(loc.text)
    except NoSuchElementException:
        job_location.append('-')
    
    #fetching job company name
    try:
        cname=driver.find_element_by_xpath('//a[@class="pad-rt-8"]')
        company_name.append(cname.text)
    except NoSuchElementException:
        company_name.append('-')
    
    #fetching job full description
    try:
        disc=driver.find_element_by_xpath('//section[@class="job-desc"]')
        full_job_description.append(disc.text)
    except NoSuchElementException:
        full_job_description.append('-')
```

```python
len(job_title),len(job_location),len(company_name),len(full_job_description)
```

```python
#Make a data frame for first 10 jobs
Data_Scientist_jobs=pd.DataFrame({})
Data_Scientist_jobs['Job-Title']=job_title
Data_Scientist_jobs['Job-Location']=job_location
Data_Scientist_jobs['Company-Name']=company_name
Data_Scientist_jobs['Job_Description']= full_job_description 
```

```python
Data_Scientist_jobs
```

```python
driver.close()
```

## Task5

#### write a python program to scrape the salary data for Data Scientist designation in Noida location. 
#### To scrape Company name, number of salaries,Average salary,Min salary,Max Salary from https://www.glassdoor.co.in/Salaries/index.htm
#### Store the collected data in a dataframe

```python
driver=webdriver.Chrome(r'C:\Users\home\Downloads\chromedriver_win32\chromedriver.exe')
```

```python
# get url
driver.get('https://www.glassdoor.co.in/Salaries/index.htm')
```

```python
#xpath to find job in search bar
search_job=driver.find_elements_by_id('KeywordSearch')[0]

#sending keys to find jobs in seach bar
search_job.send_keys('Data Scientist')

#xpath to find job location in searchbar
search_loc=driver.find_elements_by_id('LocationSearch')[0]

#sending keys to find job location in search bar
search_loc.send_keys('Noida')

#xpath to search button
search_btn=driver.find_element_by_id('HeroSearchButton')
#clicking search button
search_btn.click()
```

```python
#xpath to find company names
com_name=driver.find_elements_by_xpath('//p[@class="m-0 "]')
#fetching company names
company=[]
for c in com_name[:10]:
    company.append(c.text)

#xpath to find the no. of salaries
no_sal =driver.find_elements_by_xpath('//p[@class="css-1uyte9r css-1kuy7z7 m-0 "]')
#fetching no. of salaries
n_sal=[]
for n in no_sal[:10]:
    n_sal.append(n.text[:11])

#xpath to find the Avg salaries in company
avg_sal=driver.find_elements_by_xpath('//strong')
#fetching avarage salaries
Avg=[]
for a in avg_sal[1:20]:
    Avg.append(a.text)
res_avg = [Avg[i] for i in range(len(Avg)) if i % 2 != 1]
```

```python
#xpath to find salary range
salary=driver.find_elements_by_xpath('//p[@class="d-block d-md-none m-0 css-1kuy7z7"]')

#fetching the min salary from salary range
min_sal=[]
for i in salary[:10]:
    min_sal.append(i.text[6:15].replace('-',''))

#fetching the max salary from salary range
max_sal=[]
for i in salary[:10]:
    max_sal.append(i.text[15::].replace('-',''))
```

```python
len(company),len(n_sal),len(res_avg),len(min_sal),len(max_sal)
```

```python
#storing the data in a dataframe
Salary_details=pd.DataFrame({})
Salary_details['Comapny_name']=company
Salary_details['No.of_salaries']=n_sal
Salary_details['Avarage_salary']=res_avg
Salary_details['Min_Salary']=min_sal
Salary_details['Max_Salary']=max_sal
```

```python
#findind by id to search bar 
search=driver.find_elements_by_id('twotabsearchtextbox')[0]
#sending keys to search Laptop
search.send_keys('Laptop')
```

```python
#finding search button
search_button=driver.find_elements_by_id('nav-search-submit-button')[0]
#clicking search button
search_button.click()
```

```python
#defining attributes
Title=[]
Rating=[]
Price=[]
```

```python
#applying filter cpu type intel core i9
i9_filter=driver.find_element_by_xpath('//*[@id="p_n_feature_thirteen_browse-bin/16757432031"]/span/a/div/label/i')
i9_filter.click()
```

```python
#cpu type Intel core i9
#xpath of title
title=driver.find_elements_by_xpath('//span[@class="a-size-medium a-color-base a-text-normal"]')
#loop to get text from xpath and appending into Title list
for t in title[:10]:
    Title.append(t.text)

#xpath of price 
pri=driver.find_elements_by_xpath('//span[@class="a-price-whole"]')
#loop to get text from xpath and appending into Price list
for p in pri[:10]:
    Price.append(p.text)
#xpath of rating
rating=driver.find_elements_by_xpath('//div[@class="a-row a-size-small"]/span[1]')
#loop to get text from xpath and appending into Rating list
for r in rating[:10]:
    Rating.append(r.get_attribute("aria-label"))
```
