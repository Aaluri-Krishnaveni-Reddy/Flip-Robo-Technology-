Q1. Scrape the details of most viewed videos on youTube  from wikipedia:
    url=https://en.wikipedia.org/wiki/List_of_mostviewed_YouTube_videos
    you need to find following details:
    1.Rank
    2.Name
    3.Artist
    4.Upload date
    5.Views

```python
# opening driver 
driver=webdriver.Chrome(r'C:\Users\home\Downloads\chromedriver_win32\chromedriver.exe')

url='https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
# get url
driver.get(url)
```

```python
most_viewed={}
most_viewed['Rank']=[]
most_viewed['Name']=[]
most_viewed['Artist']=[]
most_viewed['Upload_date']=[]
most_viewed['Views']=[]

rank=driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[1]') #xpath to find rank of video
for r in rank:
    most_viewed['Rank'].append(r.text.replace('.',''))
name=driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[2]')  #xpath to find name of video
for n in name:
    most_viewed['Name'].append(n.text)
artist=driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[3]')  #xpath to find artist name 
for a in artist:
    most_viewed['Artist'].append(a.text)
date=driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[5]')  #xpath to find video upload date
for d in date[0:30]:
    most_viewed['Upload_date'].append(d.text)
    
views=driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[4]') #xpath to find views 
for v in views:
    most_viewed['Views'].append(v.text)
```

```python
len(most_viewed['Rank']),len(most_viewed['Name']),len(most_viewed['Artist']),len(most_viewed['Upload_date']),len(most_viewed['Views'])
```

```python
most_viewed=pd.DataFrame.from_dict(most_viewed)
most_viewed
```

```python
driver.close()
```

Q5.Scrape the details of tranding repositories on Github.com
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used
Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
#opening the driver 
driver=webdriver.Chrome(r'C:\Users\home\Downloads\chromedriver_win32\chromedriver.exe')
```

```python
#get url
url='https://github.com/'
driver.get(url)

#opening explore menu
explore=driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary') #xpath to find explore option
explore.click()
time.sleep(2)

#opening trending option 
trending=driver.find_element_by_xpath('//li[@class="edge-item-fix"]//following::a[@href="/trending"]')  #xpath to find trending option 
trending.click()
time.sleep(2)
```

```python
#emptylist
Github_details={}
Github_details['Repository_user']=[]                                
Github_details['Repository_Title']=[]
Github_details['Repository_description']=[]
Github_details['Contributors_count']=[]
Github_details['Language_used']=[]

#repository user name and repository title
title=driver.find_elements_by_xpath('//h1[@class="h3 lh-condensed"]/a') #xpath to extract repository title
for t in title:
    Github_details['Repository_user'].append(t.text.split("/")[0])   #appending user name in list
    Github_details['Repository_Title'].append(t.text.split("/")[1])  #appending title of repository 

#countributers
contribut_count=driver.find_elements_by_xpath('//span[@class="d-inline-block mr-3"]')  #xpath to find repository countributers count
for i in contribut_count:
    Github_details['Contributors_count'].append(len(i.find_elements_by_tag_name("a")))

    
page=[]
pages=driver.find_elements_by_xpath('//h1[@class="h3 lh-condensed"]/a')  # xpath to extract repository pages
for p in pages:
    page.append(p.get_attribute('href')) #getting attribute of repository pages

for i in page:
    driver.get(i)
    try:
        des=driver.find_element_by_xpath('//p[@class="f4 mb-3"]')  #xpath to find repository description
        Github_details['Repository_description'].append(des.text)
    except NoSuchElementException:
        Github_details['Repository_description'].append('--') 
    
    title=driver.find_elements_by_xpath('//span[@class="color-text-primary text-bold mr-1"]') #xpath to find the languages used in repository 
    lang=[]
    lang.clear()
    for t in title:
        lang.append(t.text)  
    Github_details['Language_used'].append(lang)
```

```python
len(Github_details['Repository_user']),len(Github_details['Repository_Title']),len(Github_details['Repository_description']),len(Github_details['Contributors_count']),len(Github_details['Language_used'])
```

```python
Github_details=pd.DataFrame.from_dict(Github_details)
Github_details
```

```python
driver.close()
```

Q7.Scrape the details of Data science recruiters from naukri.com.

You have to find the following details:
A) Name
B) Designation
C) Company
D) Skills they hire for
E) Location

Note: - From naukri.com homepage click on the recruiters option and the on the search pane type Data science and
click on search. All this should be done through code

```python
# opening driver 
driver=webdriver.Chrome(r'C:\Users\home\Downloads\chromedriver_win32\chromedriver.exe')
#get url
url='https://www.naukri.com/'
driver.get(url)

# click on "Recruiters"
driver.find_element_by_xpath("//ul[@class='midSec menu']//following::a[@title='Search Recruiters']").click()
time.sleep(4) # give time to load new tab
# switch to new tab
driver.switch_to.window(driver.window_handles[1])
```

```python
#search bar
searchbar=driver.find_element_by_xpath('//*[@id="skill"]/div[1]/div[2]/input')
searchbar.clear()
searchbar.send_keys("Data Scientist")

#search button
searchbutton=driver.find_element_by_id('qsbFormBtn')
searchbutton.click()

#empty lists
Recruiters_Details={}
Recruiters_Details['Name']=[]
Recruiters_Details['Designation']=[]
Recruiters_Details['Company']=[]
Recruiters_Details['Skills_they_hire']=[]
Recruiters_Details['Location']=[]

vcard_wbe=driver.find_elements_by_xpath("//div[@class='vcard']")

for i in vcard_wbe:
#     pdb.set_trace()
    if len(i.text.split('\n')) == 4: # if location info is avaialble
        Recruiters_Details['Name'].append(i.text.split('\n')[0])  #appending recuriter name 
        
        Recruiters_Details['Designation'].append(i.text.split('\n')[1])  #appending recuriter Designation
        
        Recruiters_Details['Company'].append(i.text.split('\n')[2]) #appending recuriter company
        
        Recruiters_Details['Location'].append(i.text.split('\n')[3])#appending company location
        
    elif len(i.text.split('\n')) == 3: # if location info is missing
        
        Recruiters_Details['Name'].append(i.text.split('\n')[0])
        
        Recruiters_Details['Designation'].append(i.text.split('\n')[1])
        
        Recruiters_Details['Company'].append(i.text.split('\n')[2])
        
        Recruiters_Details['Location'].append('---')

#skills
skill=driver.find_elements_by_xpath('//div[@class="hireSec highlightable"]') #xpath to find skill
for s in skill:
    Recruiters_Details['Skills_they_hire'].append(s.text)
```

Q8. Scrape the details of Highest selling novels.

compare/

You have to find the following details:
A) Book name
B) Author name
C) Volumes sold
D) Publisher
E) Genre

```python
# opening driver 
driver=webdriver.Chrome(r'C:\Users\home\Downloads\chromedriver_win32\chromedriver.exe')

# get url
url='https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare'
driver.get(url)

#emptylist
book_name=[]
author_name=[]
volume=[]
publisher=[]
genre=[]

for i in range(0,100):
    bname=driver.find_elements_by_id('table-cell-10943-{}-1'.format(i)) #xpath to find the book name
    for b in bname:
        book_name.append(b.text)
    
    aname=driver.find_elements_by_id('table-cell-10943-{}-2'.format(i)) #xpath to find the author name
    for a in aname:
        author_name.append(a.text) 
    
    vol=driver.find_elements_by_id('table-cell-10943-{}-3'.format(i)) #xpath to find volume sales
    for v in vol:
        volume.append(v.text)
        
    pub=driver.find_elements_by_id('table-cell-10943-{}-4'.format(i))  #xpath to find publisher
    for p in pub:
        publisher.append(p.text)
        
    gen=driver.find_elements_by_id('table-cell-10943-{}-5'.format(i))  #xpath to find genre
    for g in gen:
        genre.append(g.text)
```

```python
len(book_name),len(author_name),len(volume),len(publisher),len(genre)
```

```python
Highest_saled_books=pd.DataFrame({'Book_Name':book_name,'Author_Name':author_name,
                    'Volume_sales':volume,'Publisher':publisher,'Genre':genre})
Highest_saled_books
```

C) Task

```python
url='https://archive.ics.uci.edu/ml/index.php'
driver.get(url)

#opening all datasets
datasets=driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td/span/b/a')
datasets.click()

#empty lists
D_name=[]
Dtype=[]
Task=[]
A_type=[]
no_instances=[]
no_attribute=[]
Year=[]


name = driver.find_elements_by_xpath("//table[@border='1' and @cellpadding='5']/tbody/tr/td[1]/table/tbody/tr/td[2]/p[@class='normal']/b/a") #xpath to find UCI machine repositories name
for n in name:
    D_name.append(n.text)

type = driver.find_elements_by_xpath("//table[@border='1' and @cellpadding='5']/tbody/tr/td[2]/p[@class='normal']")  #xpath to find dtypes
for d in type:
    Dtype.append(d.text)

task= driver.find_elements_by_xpath("//table[@border='1' and @cellpadding='5']/tbody/tr/td[3]/p[@class='normal']") #xpath to find  task
for t in task:
    Task.append(t.text)

atype = driver.find_elements_by_xpath("//table[@border='1' and @cellpadding='5']/tbody/tr/td[4]/p[@class='normal']") #xpath to find attribute type
for a in atype:
    A_type.append(a.text)

inst = driver.find_elements_by_xpath("//table[@border='1' and @cellpadding='5']/tbody/tr/td[5]/p[@class='normal']") #xpath to find no.instance
for i in inst:
    no_instances.append(i.text)

attribute = driver.find_elements_by_xpath("//table[@border='1' and @cellpadding='5']/tbody/tr/td[6]/p[@class='normal']") #xpath to find no.attributes
for b in attribute:
    no_attribute.append(b.text)

year = driver.find_elements_by_xpath("//table[@border='1' and @cellpadding='5']/tbody/tr/td[7]/p[@class='normal']")  #xpath to find year
for y in year:
    Year.append(y.text)
```

```python
UCI_repositories=pd.DataFrame({"Dataset_name":D_name,"Data_type":Dtype,
                               "Task":Task,"Attribute_type":A_type,"No_of_instances":no_instances,
                               "No_of_attribute":no_attribute,"Year":Year})
UCI_repositories
```
